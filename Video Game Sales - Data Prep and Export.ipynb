**Data Preparation and Export**

**1. Import Pandas**

```python
import numpy as np
import pandas as pd
```

**2. Read the CSV File & Save as a Data Frame**

```python
df = pd.read_csv('vgsales.csv')

df.head()
```

**3. Data Cleaning**

**I. Drop unnecessary columns**

```python
cols = ['Rank']

df.drop(cols, inplace=True, axis=1)

df.head()
```

**II. Check for missing values**

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)

df.isnull().sum()
```

**III. Handle data types**

```python
df.dtypes
```

```python
df = df.astype({'Year':'int'})

df.dtypes
```

```python
df.head()
```

**4. DataFrame Export**

**I. Convert all rows into a list of tuples**

```python
y = []

for i in range(len(df)):
    x = tuple(df.iloc[i])
    y.append(x)
    
y
```

**II. Save the list as a txt.file**

```python
file = open('vgsales.txt', 'w', encoding='utf-8')

for tuple in y:
    file.write(str(tuple) + ',' + '\n')
    
file.close()
```
